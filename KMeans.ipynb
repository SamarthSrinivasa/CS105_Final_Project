```python
from dataloader import load_data
from sklearn.cluster import KMeans #this is the library that uses the train and test split for K Means
import numpy as np
```

```python
data = load_data('img/Thinking-of-getting-a-cat.png')
```

# creating a kmeans model with 2 clusters

```python
kmeans = KMeans(n_clusters=10, random_state=0).fit(data)

kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
kmeans.predict([[255, 255,35]])
```

```python
data[0]
```

```python
kmeans.predict([data[0]])
```

# Change each pixel to its nearest cluster

```python
for i in range(len(data)):
    predicted = kmeans.predict([data[i]])
    data[i] = kmeans.cluster_centers_[predicted]
    # print(predicted)
    # print(data[i])
```
